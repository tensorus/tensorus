## ğŸ“¦ Part 1: Creating Different Types of Tensors

Let's create tensors step by step, from simple to complex!

```python
# ğŸ”¢ 0D Tensor (Scalar) - Single number
print("ğŸ“¦ 0D TENSOR (SCALAR)")
print("=" * 30)

# Create scalars
temperature = torch.tensor(23.5)  # Temperature in Celsius
price = torch.tensor(99.99)       # Product price
accuracy = torch.tensor(0.95)     # Model accuracy

print(f"ğŸŒ¡ï¸  Temperature: {temperature} Â°C")
print(f"ğŸ’° Price: ${price}")
print(f"ğŸ¯ Accuracy: {accuracy:.1%}")
print(f"ğŸ“ Shape: {temperature.shape} â† No dimensions!")
print(f"ğŸ“Š Data type: {temperature.dtype}")

# Mathematical operations on scalars
fahrenheit = temperature * 9/5 + 32
print(f"ğŸ”¥ Temperature in Fahrenheit: {fahrenheit:.1f}Â°F")

print("\nğŸ’¡ Use cases: Single measurements, constants, final predictions")
```

```python
# ğŸ“ 1D Tensor (Vector) - Array of numbers
print("ğŸ“ 1D TENSOR (VECTOR)")
print("=" * 30)

# Create vectors for different use cases
stock_prices = torch.tensor([100.0, 105.2, 98.7, 102.1, 110.5, 108.9, 115.3])
sensor_readings = torch.tensor([23.1, 23.4, 23.2, 23.8, 24.1, 23.9, 23.6])
word_embedding = torch.randn(384)  # Typical embedding size

print(f"ğŸ“ˆ Stock prices (7 days): {stock_prices}")
print(f"ğŸ“ Shape: {stock_prices.shape} â† 7 values in sequence")
print(f"ğŸŒ¡ï¸  Sensor readings: {sensor_readings}")
print(f"ğŸ§  Word embedding shape: {word_embedding.shape} (first 5: {word_embedding[:5]})")

# Vector operations
print("\nğŸ”¢ Vector Statistics:")
print(f"   ğŸ“Š Stock price mean: ${stock_prices.mean():.2f}")
print(f"   ğŸ“ˆ Stock price max: ${stock_prices.max():.2f}")
print(f"   ğŸ“‰ Stock price min: ${stock_prices.min():.2f}")
print(f"   ğŸ“ Vector length (norm): {torch.norm(stock_prices):.2f}")

# Practical example: Calculate daily returns
daily_returns = (stock_prices[1:] - stock_prices[:-1]) / stock_prices[:-1] * 100
print(f"\nğŸ’¹ Daily returns (%): {daily_returns}")

print("\nğŸ’¡ Use cases: Time series, embeddings, feature vectors, sensor data")
```
